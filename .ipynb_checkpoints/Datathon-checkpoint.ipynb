```python
def gam(a, b, k, maxim):
    sim = []
    i = 1
    while i<=k:
        g = np.rint(np.random.gamma(a, 1/b, 1)[0])
        if g<=maxim:
            sim.append(g)
            i = i+1
    return sim
```

```python
data = train1[feature].dropna().to_numpy().reshape(-1, 1)
g = mixture.GaussianMixture(n_components = 4, random_state=0, max_iter=200).fit(data)
xpdf = np.linspace(min(data), max(data), len(train1[feature]))
density = (np.exp(g.score(xpdf)))
```

```python
for i in range(g.n_components):
    pdf = g.weights_[i]*stats.norm(g.means_[i,0],
                                  np.sqrt(g.covariances_[i,0])).pdf(xpdf)
    plt.fill(xpdf, pdf, facecolor = 'gray', edgecolor = 'none', alpha = 0.5)
    sns.kdeplot(train1[feature], color = '#ffd514')
    plt.hist(train1[feature], density = True, bins = 25, alpha = 0.05, color = 'b')
    plt.title('State_1')
```

```python
#Encontrar el número óptimo de clústers basados en BIC
x = data
bics = []
min_bic = 0
counter=1
for i in range (10):
  gmm = GMM(n_components = counter, max_iter=len(train1[feature]), random_state=0, covariance_type = 'full')
  labels = gmm.fit(x).predict(x)
  bic = gmm.bic(x)
  bics.append(bic)
  if bic < min_bic or min_bic == 0:
    min_bic = bic
    opt_bic = counter
  counter = counter + 1

fig = plt.figure(figsize=(8, 3))
plt.plot(np.arange(1,11), bics, 'o-', lw=3, c='black', label='BIC')
plt.legend(frameon=False, fontsize=15)
plt.xlabel('Número de clústers', fontsize=20)
plt.title('Clúters óptimos = '+str(opt_bic), fontsize=20)
plt.show()
```

```python
g_mixture = (g.weights_[0]*stats.norm(g.means_[0,0], np.sqrt(g.covariances_[0,0])).pdf(xpdf)+ 
             g.weights_[1]*stats.norm(g.means_[1,0], np.sqrt(g.covariances_[1,0])).pdf(xpdf)+
             g.weights_[2]*stats.norm(g.means_[2,0], np.sqrt(g.covariances_[2,0])).pdf(xpdf)+
             g.weights_[3]*stats.norm(g.means_[3,0], np.sqrt(g.covariances_[3,0])).pdf(xpdf))
```

```python
def sim_mixture(k):
    sim_mix = (g.weights_[0]*np.random.normal(g.means_[0,0],np.sqrt(g.covariances_[0,0]), k)+
              g.weights_[1]*np.random.normal(g.means_[1,0],np.sqrt(g.covariances_[1,0]), k)+
              g.weights_[2]*np.random.normal(g.means_[2,0],np.sqrt(g.covariances_[2,0]), k)+
              g.weights_[3]*np.random.normal(g.means_[3,0], np.sqrt(g.covariances_[3,0]), k))
    return sim_mix
```
